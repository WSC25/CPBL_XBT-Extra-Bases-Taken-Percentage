## 計算各個進壘狀況

```python
# 跑者在一壘時 打者擊出一壘安打
first_1B = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 1)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='當跑者在一壘_擊出1B')

#跑者在一壘時 當打者擊出一壘安打時 跑者推進至二壘
first_S_2 = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 1) & (merged_df['結束壘包'] == 2)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='1B_1壘跑者推進至2壘次數')

#跑者在一壘時 當打者擊出一壘安打時 跑者推進至三壘
first_S_3 = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 1) & (merged_df['結束壘包'] == 3)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='1B_1壘跑者推進至3壘次數')

#跑者在一壘時 當打者擊出一壘安打時 跑者回到本壘
first_S_H = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 1) & (merged_df['結束壘包'] == 0)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='1B_1壘跑者回到本壘次數')

# 跑者在一壘時 打者擊出二壘安打
first_2B = merged_df[(merged_df['打席結果'] == '2B') & (merged_df['開始壘包'] == 1)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='當跑者在一壘_擊出2B')

#跑者在一壘時 當打者擊出二壘安打時 跑者推進至三壘
first_D_3 = merged_df[(merged_df['打席結果'] == '2B') & (merged_df['開始壘包'] == 1) & (merged_df['結束壘包'] == 3)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='2B_1壘跑者推進至3壘次數')

#跑者在一壘時 當打者擊出二壘安打時 跑者直接回壘
first_D_H = merged_df[(merged_df['打席結果'] == '2B') & (merged_df['開始壘包'] == 1) & (merged_df['結束壘包'] == 0)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='2B_1壘跑者回本壘次數')

# 跑者在二壘時 打者擊出一壘安打
second_2B = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 2)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='當跑者在二壘_擊出1B')

#跑者在二壘時 當打者擊出一壘安打時 跑者推進至三壘
second_S_3 = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 2) & (merged_df['結束壘包'] == 3)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='1B_2壘跑者推進至3壘次數')

#跑者在二壘時 當打者擊出一壘安打時 跑者直接回壘
second_S_H = merged_df[(merged_df['打席結果'] == '1B') & (merged_df['開始壘包'] == 2) & (merged_df['結束壘包'] == 0)].groupby(['球隊','跑者'])['打席結果'].count().reset_index(name='1B_2壘跑者回本壘次數')
```

## 合併各跑者進壘狀況+計算XBT%

```python
#所有結果合併
group_columns = ['球隊','跑者']
df1 = pd.merge(first_1B,first_S_2, on=group_columns, how='outer')
df2 = pd.merge(df1,first_S_3, on=group_columns, how='outer')
df3 = pd.merge(df2,first_S_H, on=group_columns, how='outer')
df4 = pd.merge(df3,first_2B, on=group_columns, how='outer')
df5 = pd.merge(df4,first_D_3, on=group_columns, how='outer')
df6 = pd.merge(df5,first_D_H, on=group_columns, how='outer')
df7 = pd.merge(df6,second_2B, on=group_columns, how='outer')
df8 = pd.merge(df7,second_S_3, on=group_columns, how='outer')
result = pd.merge(df8,second_S_H, on=group_columns, how='outer')
result = result.fillna(0)
```

```python
result.info()
```

```python
result['XBT%'] =(result['1B_1壘跑者推進至3壘次數']+result['2B_1壘跑者回本壘次數']+result['1B_2壘跑者回本壘次數']+result['1B_1壘跑者回到本壘次數'])/(result['當跑者在一壘_擊出1B']+result['當跑者在一壘_擊出2B']+result['當跑者在二壘_擊出1B'])
```

## 合併球隊 跑者進壘狀況+計算球隊XBT%

```python
#球隊各個狀況合併
team_first_1B=first_1B.groupby(['球隊']).agg({'當跑者在一壘_擊出1B':'sum'})
team_first_S_2=first_S_2.groupby(['球隊']).agg({'1B_1壘跑者推進至2壘次數':'sum'})
team_first_S_3=first_S_3.groupby(['球隊']).agg({'1B_1壘跑者推進至3壘次數':'sum'})
team_first_S_H=first_S_H.groupby(['球隊']).agg({'1B_1壘跑者回到本壘次數':'sum'})
team_first_2B=first_2B.groupby(['球隊']).agg({'當跑者在一壘_擊出2B':'sum'})
team_first_D_3=first_D_3.groupby(['球隊']).agg({'2B_1壘跑者推進至3壘次數':'sum'})
team_first_D_H=first_D_H.groupby(['球隊']).agg({'2B_1壘跑者回本壘次數':'sum'})
team_second_2B=second_2B.groupby(['球隊']).agg({'當跑者在二壘_擊出1B':'sum'})
team_second_S_3=second_S_3.groupby(['球隊']).agg({'1B_2壘跑者推進至3壘次數':'sum'})
team_second_S_H=second_S_H.groupby(['球隊']).agg({'1B_2壘跑者回本壘次數':'sum'})
```

```python
group_columns = ['球隊']
df1 = pd.merge(team_first_1B,team_first_S_2, on=group_columns)
df2 = pd.merge(df1,team_first_S_3, on=group_columns)
df3 = pd.merge(df2,team_first_S_H, on=group_columns)
df4 = pd.merge(df3,team_first_2B, on=group_columns)
df5 = pd.merge(df4,team_first_D_3, on=group_columns)
df6 = pd.merge(df5,team_first_D_H, on=group_columns)
df7 = pd.merge(df6,team_second_2B, on=group_columns)
df8 = pd.merge(df7,team_second_S_3, on=group_columns)
team_result = pd.merge(df8,team_second_S_H, on=group_columns)
```

```python
team_result['XBT%'] =(team_result['1B_1壘跑者推進至3壘次數']+team_result['2B_1壘跑者回本壘次數']+team_result['1B_2壘跑者回本壘次數']+team_result['1B_1壘跑者回到本壘次數'])/(team_result['當跑者在一壘_擊出1B']+team_result['當跑者在一壘_擊出2B']+team_result['當跑者在二壘_擊出1B'])
```
