## 主客場合併

```python
merged_df = pd.concat([merged_home, merged_away], ignore_index=True)
```

```python
merged_df
```

```python
merged_df['XBT'].value_counts()
```

```python
#將額外進壘"2次" 改為1 表示"有額外進壘"
#XBT = "-1" 表示"沒有額外進壘"
merged_df['XBT'] = merged_df['XBT'].replace(2, 1)
merged_df['XBT'] = merged_df['XBT'].replace(-1, 0)
```

```python
merged_df['XBT'].value_counts()
```

## 計算球員個人XBT%

```python
dfs = {}

# 定義你想分組的欄位
group_columns = ['跑者']

# 分組和聚合計算出局數的次數和總和
for name, group in merged_df.groupby(group_columns):
    df1 = merged_df.groupby('跑者').size().reset_index(name='跑者資料筆數')
    df2 = group.groupby(group_columns).agg({'XBT':'sum'})
    dfs[name] = pd.merge(df1, df2, on=group_columns)

# 合併所有的 DataFrame
result = pd.concat(dfs.values())
result = result.rename(columns={'XBT': '有額外進壘次數'})
result
```

```python
result['XBT%'] = result['有額外進壘次數']/result['跑者資料筆數']*100
result
```

```python
dfs = {}

# 定義你想分組的欄位
columns = ['球隊']

for name, group in merged_df.groupby(columns):
    df1 = merged_df.groupby('球隊').size().reset_index(name='跑者資料筆數')
    df2 = group.groupby(columns).agg({'XBT':'sum'})
    dfs[name] = pd.merge(df1, df2, on=columns)

# 合併所有的 DataFrame
result2 = pd.concat(dfs.values())
result2 = result2.rename(columns={'XBT': '有額外進壘次數'})
result2['XBT%'] = result2['有額外進壘次數']/result2['跑者資料筆數']*100
result2
```
